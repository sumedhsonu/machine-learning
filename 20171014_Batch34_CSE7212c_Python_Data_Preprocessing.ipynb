```python
def summaryx(x):

     a=x.mean()
     b=x.median()
     c=x.std()
     d=x.quantile(0.25)
     e=x.quantile(0.75)
     summary_x=pd.DataFrame([a,b,c,d,e],index=['mean','median','std','1-quantile','3-quantile'])
     return summary_x;
```
