```python
#importing libraries
import numpy as np
import pandas as pd
import os
```

```python
#reading train dataset
df=pd.read_csv("train.csv")
```

```python
df
```

```python
#dropping rows which has na values
df1=df.dropna()
```

```python
#sending column label column to y
y = df1.label
```

```python
y
```

```python
#splitting the data into train and test for model building
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df1['text'], y, test_size=0.33, random_state=53)
```

```python
#reading test data
df2=pd.read_csv("test.csv")
```

```python
#dropping na values
df3=df2.dropna()
```

```python
#taking only text column from test data
xtest1=df3["text"]
xtest1
```

```python
#counter vectorization and fitting the data 
from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer = CountVectorizer(stop_words='english')
count_train = count_vectorizer.fit_transform(X_train)
count_test = count_vectorizer.transform(X_test)
count_test1 = count_vectorizer.transform(xtest1)
```

```python
#tfid vectorization method
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_df=0.5)
tfidf_train = tfidf_vectorizer.fit_transform(X_train)
tfidf_test = tfidf_vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
```

```python
#building the model with naive bayes classification
clf = MultinomialNB()
clf.fit(count_train, y_train)
pred = clf.predict(count_test)
score = metrics.accuracy_score(y_test, pred)
print("accuracy:   %0.3f" % score)
cm = metrics.confusion_matrix(y_test, pred,)
```
